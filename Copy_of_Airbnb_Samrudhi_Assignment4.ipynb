### Mean, Median, Mode Tangent

```python
# from the dataset i have chosen age as an attribute to find mean mode and median
import urllib.request

Ava_url = 'https://raw.githubusercontent.com/Samrudhig/datasets/main/availability_airbnb'
Ava = []
for line in urllib.request.urlopen(Ava_url):
    Ava.append(line.decode('utf-8'))

Ava_data = [int(x.rstrip()) for x in Ava] # remove line breaks
print(Ava_data)
print(len(Ava_data))
```

```python
# mean - add everything up and divide by n
sum(Ava_data)/len(Ava_data)
```

```python
# median - sort everthing and then find the middle value
Ava_data.sort()
print(Ava_data)
print(Ava_data[0])
print(len(Ava_data)/2)
print(Ava_data[int(len(Ava_data)/2)])
```

```python
# mode - find the most frequent occurence
most_frequent = {}
for i in Ava_data:
  if i in most_frequent:
   most_frequent[i] += 1
  else:
    most_frequent[i] = 1
max(most_frequent, key=most_frequent.get)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

xmean = np.mean(Ava_data)
xmedian = np.median(Ava_data)
xmode = stats.mode(Ava_data, keepdims=True)[0][0]

fig = plt.figure(figsize=(10,5))
sns.histplot(Ava_data, kde=True, color='black')
plt.axvline(x=xmean, color='purple', label='Mean', linewidth=8)
plt.axvline(x=xmedian, color='blue', label='Median', linewidth=5)
plt.axvline(x=xmode, color='green', label='Mode')
plt.legend()
plt.show()
```

```python
# standard normal distribution with numpy
# https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html
mu = 0
sigma = 1
x = np.random.normal(0, 1, 1000)
count, bins, ignored = plt.hist(x, density=True, bins=25)

print('Variance:', x.var(ddof=0))

# plot the distribution curve
plt.plot(bins, 1/(np.sqrt(2 * np.pi * sigma**2)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ), linewidth=3, color='k')
plt.xlabel('Standard Deviations')
plt.ylabel('Probabilities')
plt.axvline(x=mu, color='red', label='Mean')
plt.title('Normal Standard Distribution')
plt.legend()
plt.show()
```

```python
# visualize outliers before and after
from random import random
from random import randint
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise=50, random_state=42)
print('mean before outliers:', np.mean(X))
print('var before outliers:', np.var(X))
fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(2, 2, figsize=(10, 10))
ax1.scatter(X, y)
ax2.hist(X)
ax3.boxplot(X)
sns.violinplot(ax=ax4, data=X);
```

```python
# create outliers
for i in range(30):
    factor = randint(-3, 3)
    if random() > 0.5:
        X[i] += factor * X.std()
    else:
        X[i] -= factor * X.std()

fig, ([ax1, ax2], [ax3, ax4]) = plt.subplots(2, 2, figsize=(10, 10))
ax1.scatter(X, y)
ax2.hist(X)
ax3.boxplot(X)
sns.violinplot(ax=ax4, data=X)
print('mean after outliers:', np.mean(X))
print('var after outliers:', np.var(X))
```

```python
import numpy as np
from scipy.stats import skewnorm
from scipy import stats
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 2, sharey=True, figsize=(14, 3))
skew = 10
n = 100000

r = skewnorm.rvs(skew, loc=0, scale=10, size=n)
ax[0].hist(r, bins=50, density=True, alpha=0.4)
ax[0].axvline(x=np.median(r), color='blue', label='Median')
ax[0].axvline(x=np.mean(r).round(2), color='black', label='Mean')
ax[0].legend()
print(f'right skew data mean: {np.mean(r).round(2)}, median: {np.median(r).round(2)}')

l = skewnorm.rvs(-skew, loc=0, scale=10, size=n)
ax[1].hist(l, bins=50, density=True, alpha=0.4);
ax[1].axvline(x=np.mean(l).round(2), color='blue', label='Mean')
ax[1].axvline(x=np.median(l), color='black', label='Median')
ax[1].legend()
print(f'left skew data mean: {np.mean(l).round(2)}, median: {np.median(l).round(2)}')
```
